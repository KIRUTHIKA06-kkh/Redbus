```python
import pandas as pd
import numpy as np
```

All state details

```python
df_ap=pd.read_csv('all_bus_details_ap.csv')
df_k=pd.read_csv('all_bus_details_k.csv')
df_t=pd.read_csv('all_bus_details_t.csv')
df_kad=pd.read_csv('all_bus_details_kad.csv')
df_r=pd.read_csv('all_bus_details_r.csv')
df_sb=pd.read_csv('all_bus_details_sb.csv')
df_hp=pd.read_csv('all_bus_details_hp.csv')
df_a=pd.read_csv('all_bus_details_a.csv')
df_up=pd.read_csv('all_bus_details_up.csv')
df_b=pd.read_csv('all_bus_details_b.csv')
```

```python
redbus_df=pd.concat([df_ap,df_k,df_t,df_kad,df_r,df_sb,df_hp,df_a,df_up,df_b],ignore_index=True)
redbus_df
```

```python
redbus_df.info()
```

```python
redbus_df["Price"]=redbus_df["Price"].str.replace("INR","")
redbus_df["Price"]=redbus_df["Price"].astype(float)
redbus_df["Price"].fillna(0)
```

```python
redbus_df.info()
```

```python
redbus_df=redbus_df.replace({np.nan:None})
```

```python
redbus_df.to_csv('redbus_df.csv',index=False)
```
